```python
from download_yt_video import download_youtube_by_id

# Example usage
# youtube_url = "https://www.youtube.com/watch?v=jANZxzetPaQ"
youtube_url = "https://www.youtube.com/watch?v=pveIuZT0GJE"
video_path = download_youtube_by_id(youtube_url)
```

```python
from mavi_pipeline import (
    upload_video, get_video_metadata,
    submit_transcription, get_transcription
)

# 1. Upload
upload_info = upload_video(video_path)
video_no = upload_info["videoNo"]
print(f"‚úÖ Uploaded: {video_no}")
```

```python
# 2. Check metadata (status should be PARSE before proceeding)
import time
meta = get_video_metadata(video_no)
print(f"‚úÖ Metadata: {meta}")
```

```python
# 3. Submit transcription (AUDIO + VIDEO)
audio_res = submit_transcription(video_no, "AUDIO")
audio_task_no = audio_res["data"]["taskNo"]
# 4. Poll results (can wrap this in retry logic if needed)
time.sleep(15)  # wait before retrieving
audio_result = get_transcription(audio_task_no)
# 5. Print transcripts
print("üéß AUDIO TRANSCRIPT:")
for t in audio_result["data"].get("transcriptions", []):
    print(f"{t['startTime']}s - {t['endTime']}s:\n{t['content']}\n")
```

```python
# 3. Submit transcription (VIDEO)
video_res = submit_transcription(video_no, "VIDEO")
video_task_no = video_res["data"]["taskNo"]
time.sleep(60)  # wait before retrieving
video_result = get_transcription(video_task_no)
print("üéûÔ∏è VIDEO TRANSCRIPT:")
for t in video_result["data"].get("transcriptions", []):
    print(f"{t['startTime']}s - {t['endTime']}s:\n{t['content']}\n")
```
